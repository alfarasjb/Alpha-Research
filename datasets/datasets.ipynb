### **FETCHING DATA FROM ALPHA VANTAGE** 

```python
import pandas as pd 
from alpha_vantage.timeseries import TimeSeries 
import os 
import mplfinance as mpf
import matplotlib.pyplot as plt

key = os.environ['ALPHA_VANTAGE_API_KEY']
ts = TimeSeries(key=key, output_format='pandas',indexing_type='date')


SYMBOL = "BTCUSD"

data=ts.get_daily(SYMBOL,outputsize='full')

df=data[0][::-1]
df.columns = ['open','high','low','close','volume']
df.head()

```

```python
mpf.plot(df.tail(200),type='candle',volume=True, title="BTCUSD - OHLCV Last 200 Days", figsize=(10, 4))
```
