## TRADETESTLIB DEMO

This notebook is a demonstration of [`tradetestlib`](https://github.com/alfarasjb/TradeTestLib/tree/main) utilizing one of my strategies in development. 


The objective of this notebook is to assess a basket of currencies, more specifically, Forex Majors, regarding their performance with the strategy in test, and to determine it is profitable to deploy in live market conditions. 

**Brief Strategy Description:**

The strategy takes top `n_samples` best performing time intervals, and converts it to signals to use for trading. 



**Limitations:**
1. Data Availability - since the data source is raw broker data on the 5 minute timeframe, testing period, in terms of years is shortened 
2. Trading Flexibility - the code below only provides a broad outlook as to how the strategy would perform. In a more realistic sense, it would be best to optimize parameters further when building the execution on MQL4, such as applying trailing stops. 
3. Spread Limit - assessing the impact of spread to overall performance may vary since bid-ask spread varies during different times of day, hence it poses a challenge to accurately determine the spread. For this test, spread will be taken from live market conditions during the time of testing. However, a helper function is also created to assess the performance of the strategy at different spreads, therefore disabling the algo when the spread is not favorable, and determining the spread at which the algo performance degrades significantly. 


**Evaluation Metrics|** 
1. Evaluation metrics will be found in the `evaluation.py` file. For this test, `net_profit`, `profit_factor`, and `expectancy` will be more in focus. 


**Optimization|**
1. The algo will be optimized by a grid search algorithm, and by maximizing, or minimizing evaluation metrics. For this test, Optimization will be done by maximizing `profit_factor`, calculated as| 

```
Profit Factor = (WR * Pa) / (LR * La)

Where| 
Pa = Average Profit
La = Average Loss 
WR = Win Rate 
LR = Loss Rate

```

General Purpose Helper Functions for more efficient backtesting

1. `test_dataframe_processing`: validates entire backtest dataframe for any miscalculated parameters, stray profits, stray losses. 
2. `get_data`: helper function for generating train_data, test_data
3. `run`: runs a simulation. `evaluation` and `summary` may be accessed 
4. `simulate_group`: takes a batch of currencies from MT5, and runs a simulation on similar configurations, and identifying instruments that respond well to the algo (training set)
5. `run_optimization`: optimizes a simulation based on tunable hyperparameters, and by utilizing a grid search algo 
6. `get_spread_limit`: performs an analysis as to the minimum spread that makes the strategy performance deteriorate

````python
## HELPER FUNCTIONS 

# UTILITY METHODS

def test_dataframe_processing(s: Simulation, raw: pd.DataFrame):
    """"""
    
    assert len(raw.query('signal != 0 & trade_diff == 0 & net_profit > 0')) == 0
    assert len(raw.query('raw_profit == 0 & spread_adjusted_profit > 0 & signal != 0')) == 0
    assert len(raw.query('raw_profit < 0 & raw_profit < spread_adjusted_profit')) == 0
    assert len(raw.query('spread_adj_trade_diff < 0 & net_profit > 0')) == 0
    
    assert len(raw.query('signal == 0 & net_profit < 0')) == 0
    assert len(raw.query('signal == 0 & net_profit > 0')) == 0
    assert len(raw.query('raw_profit < 0 & raw_profit < spread_adjusted_profit')) == 0

def get_data(symbol:str , tf: str, num_samples: int):
    arb = Strat(symbol, tf, 0.2, num_samples, calculate_by = 'mean')
    train_data, test_data, sig_map, signals_reference = arb.build_dataframe()
    return train_data, test_data, sig_map


def run(symbol: str, 
        tf: str, 
        train_data: pd.DataFrame, 
        test_data: pd.DataFrame,
        starting_balance: int = 100000,
        lot: float = 1,
        holdtime: int = 5, 
        max_loss: float = 0.01, 
        show_properties:bool = False,
        spread = None):

    sim = Simulation(symbol = symbol, 
                     timeframe = tf, 
                     train_raw = train_data, 
                     test_raw = test_data, 
                     lot = lot, 
                     starting_balance = starting_balance, 
                     hold_time = holdtime, 
                     show_properties = show_properties, 
                     commission = 3.5,
                     max_loss_pct = max_loss, 
                     spread = spread)
    
    return sim

def simulate_group(group:str = 'Forex', tf:str = 'm5', sort_by:str = 'profit_factor', samples:int = 10):
    
    mt5_data = MT5_Data()
    symbols_list = mt5_data.get_mt5_symbols(group)
    
    processed = []
    added = []
    simulations = {}
    exclude = ['TRY','ZAR','RUB','JPY','SEK']
    basket = []
    basket_summary = pd.DataFrame()
    for m in tqdm(symbols_list):
        try:
            processed.append(m)
            train_data, test_data, sig_map = get_data(m, tf, samples)
            train_data.columns = [c.lower() for c in train_data.columns]
            test_data.columns = [d.lower() for d in test_data.columns]
            sim = run(m, tf, train_data, test_data, lot = 10, holdtime = 5, max_loss = 0.01)

            if (sim.train_filtered_evaluation.spread > 100) or (sim.train_filtered_evaluation.sharpe_ratio < 0):
                continue 
                
            basket.append(sim.train_filtered_summary)
            added.append(m)
            simulations[m] = sim
            
        except KeyError:
            print('Key Error')
            continue
        

        
        basket_summary = pd.concat(basket, axis = 1)
        basket_summary = basket_summary.sort_values(by = sort_by, ascending = False, axis = 1)
        
    print(f'{len(processed)} Symbols Processed')
    print(f'{len(added)} Symbols Added')
    print(added)
    return added, simulations, basket_summary


def run_optimization(symbol: str, 
                     tf: str,
                     samples = 10, 
                     starting_balance = 100000,
                     metric: str = 'profit_factor', 
                     how: str = 'maximize', 
                     lot_range = np.arange(2, 10, 2),
                     holdtime_range = [5, 10],
                     max_loss_range = [0.005, 0.01, 0.02],
                     dataset = 'train'):


    train, test, sigmap = get_data(symbol, tf, samples)
    train.columns = [c.lower() for c in train.columns]
    test.columns = [d.lower() for d in test.columns]
    optimize = Optimize(symbol, tf, train.copy(), test.copy(), starting_balance = starting_balance,
                        metric = metric,how = how, dataset = dataset)

    params = {
        'lot':lot_range,
        'hold_time':holdtime_range,
        'max_loss':max_loss_range,
    }
    
    best, df = optimize.run_grid_search(params)
    
    return best, optimize

def get_spread_limit(metric:str, max_spread:int, symbol:str, timeframe:str, train:pd.DataFrame, test:pd.DataFrame, lot:float, holdtime:int, max_loss:float):
    
    items = []
    for i in np.arange(1,max_spread):

        spr_test = run(symbol, timeframe, train, test, lot = lot, holdtime = holdtime, max_loss = max_loss, spread = i)
        evaluation = spr_test.test_filtered_evaluation
        prop = spr_test.test_filtered_summary.loc[metric].item()
        measured = 0 if evaluation.max_dd_pct == 100 else prop
        items.append([i,measured])

    items_df = pd.DataFrame(items)
    items_df = items_df.set_index(0)

    label = metric.replace('_', ' ').title()
    items_df.plot()
    plt.title('Spread Limit')
    plt.xlabel('Spread Ticks')
    plt.ylabel(label)
    plt.legend(labels = [label])
````

Creates a basket simulation of a group of currencies, and determines the best performing currencies by `profit_factor`

````python
mt5.initialize()
added, simulations, basket_summary = simulate_group(group = 'Majors',tf = 'm5', sort_by = 'profit_factor')
````

`basket_summary` contains a concatenated datadrame of `train_filtered` summary from the `simulate_group` method.

`train_filtered` is the summary when interval filtering is taken into account, and will also be used for optimization, and validation on the `test_filtered` dataset.

Interval filtering is performed by identifying the best time ranges in the `train_data` set in which the strategy performs best. This result will then be applied onto the `test_data`, which creates the `test_filtered` dataset, evaluation, and summary, which can then be further scrutinized in the later portions of this notebook.

````python
basket_summary
````

While USDCHF, and USDJPY perform better than AUDUSD in terms of `profit_factor`, the latter shows a far superior `win_rate`, and more manageable `max_dd_pct`. For that purpose, AUDUSD will be optimized in this test. 

1. `lot_range` is just a selection of possible lot sizes to use for the test. 
2. `holdtime_range` will be the default values, [5, 10]
3. `max_loss_range` will be 0.5%, 1%, and 2% of deposit [0.005, 0.01, 0.02]

In this case, `train_filtered` will be optimized based on `maximizing` the `profit_factor`. 

`run_optimization` returns the dataframe containing the best combination of hyperparameters specified. Additionally, the values will also be accessible as follows: 

`opt.optimized_lot`, `opt.optimized_holdtime`, `opt.optimized_max_loss`

which can then be used for validating on the `test_filtered` dataset. 

````python
best, opt = run_optimization('AUDUSD','m5',10, lot_range = [8, 10, 12], starting_balance = 100000, metric = 'profit_factor', how = 'maximize')
best
````

Optimization Results are shown in the dataframe above.

````python
au = simulations['AUDUSD']
au_opt = run('AUDUSD', 'm5', au.train_raw, au.test_raw, lot = opt.optimized_lot, holdtime = opt.optimized_holdtime, max_loss = opt.optimized_max_loss,show_properties = True)
````

**Evaluating with MetaTrader Script** 

A script on MetaTrader4 was made for the purposes of validating the results of the Python-based test, and running a tick simulation on MetaTrader4 Strategy Tester, evaluating the performance. 

````python
filtered_test = au_opt.test_filtered.loc[au_opt.test_filtered['net_profit'] != 0][['net_profit','balance']]

mt4_backtest = pd.read_csv('backtest_mt4.csv', index_col = 'Date')
mt4_backtest.index = pd.to_datetime(mt4_backtest.index, format = '%Y-%m-%d')

combined_test_set = pd.concat([filtered_test, mt4_backtest], axis = 1)
combined_test_set = combined_test_set.ffill()
combined_test_set = combined_test_set.loc[combined_test_set['sim_net_profit'] != 0]

````

**Profit Comparison: Python Backtest vs MetaTrader Backtest** 

The figure above the net profit per trade for each signal. It can be seen that the net profit figures of the MT4 test are almost exactly identical to that of the Python test.

**Equity Curve Comparison: Python Backtest vs MetaTrader Backtest** 

The figure above shows a comparison of the Equity Curve of the Python Backtest vs MetaTrader backtest. It shows, reflecting the the previous figure, that the results for both tests are almost identical.
